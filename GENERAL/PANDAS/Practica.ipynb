```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv("avocado.csv")
df["region"][:49]
chicago = df[df["region"] == "Chicago"]
chicago = chicago.set_index("Date")
chicago = chicago.sort_values(by="Date")
#Grafico
MAX_SAMPLES = 100
precio = chicago["AveragePrice"][:MAX_SAMPLES]
Cantidad = chicago["Total Volume"][:MAX_SAMPLES]
plt.plot(precio,label="Precio Medio")
plt.plot(Cantidad,label="Volumen total")
plt.title("Precio de los aguacates vs tiempo")
plt.xlabel("Fecha")
plt.xticks(rotation=90)
plt.ylabel("Precio en $")
plt.legend()
plt.show()


```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df = pd.read_csv("avocado.csv")
atlanta = df[df["region"] == "Atlanta"]
#atlanta = atlanta[:2]
#atlanta = atlanta.set_index('Date')
atlanta = atlanta.sort_index()
precio = atlanta["AveragePrice"]
precioPromedio = precio.rolling(30,min_periods =1).mean()
volumen = atlanta["Total Volume"]

bolsasAguacate = atlanta["Total Bags"]
sbolsas = atlanta["Small Bags"]
lbolsas = atlanta["Large Bags"]
xlbolsa = atlanta["XLarge Bags"]
plt.subplot(221)
plt.title("Precio aguacate")
plt.plot(precio,label="Precio",color = "blue")
plt.plot(precioPromedio,label="Precio Promediado",color = "orange")
plt.legend()
#plt.xticks(rotation=90)

plt.subplot(222)
plt.title("Volumen aguacate")
plt.plot(volumen,label="Volumen Total",color="red")
plt.legend() 

plt.subplot(223)
plt.title("Bolsas Totales de aguacates")
plt.plot(bolsasAguacate,label="Bolsas Total",color="blue")
plt.legend()

plt.subplot(224)
plt.title("Bolsas por tamaÃ±o")
plt.plot(sbolsas,label="Bolsas - S",color="black")
plt.plot(lbolsas,label="Bolsas - L ",color="cyan")
plt.plot(xlbolsa,label="Bolsas - XL",color="yellow")
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import re
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
results = pd.read_csv("survey_results_public.csv",index_col="ResponseId")
resultsSchema = pd.read_csv("survey_results_schema.csv")
languages = "LanguageWantToWorkWith"
salary = "ConvertedCompYearly"
results.rename(columns={languages:"Languages",salary:"Salary"},inplace=True)
results.dropna(subset=["Salary","Languages"],inplace=True)
results = results.loc[:,["Country","Salary","Languages"]].sort_values(by="Salary")
filtro = (results["Salary"] >= 10000) & (results["Salary"] <= 3e6)
results = results[filtro]
lg = results["Languages"]
alLanaguages = lg[55969].split(";")
for row in lg:
    for lang in row.split(";"):
        alLanaguages.append(lang)
alLanaguages = list(set(alLanaguages))
alLanaguages.sort()
meanSalary = pd.DataFrame(data=np.zeros(len(alLanaguages)),index=alLanaguages,columns=["Salary"])
for lang in alLanaguages:
    mask = results["Languages"].str.contains(re.escape(lang))
    average = results.loc[mask,"Salary"].mean()
    meanSalary.loc[lang,"Salary"] = average
colors = [
    "#03071E",
    "#370617",
    "#6A040F",
    "#9D0208",
    "#D00000",
    "#DC2F02",
    "#E85D04",
    "#F48C06",
    "#FAA307",
    "#FFBA08"
]
fontdict = {
    'family':'serif',
    'color':'darkred',
    'weight':'normal',
    'size':16
}
%matplotlib qt5

plt.style.use("seaborn")
meanSalary.sort_values(by="Salary",inplace=True)
plt.barh(meanSalary.index,meanSalary["Salary"],color=colors) 
plt.vlines(meanSalary["Salary"].mean(),0,len(meanSalary["Salary"]),color="red",linestyles="dotted")
plt.title("Programming Languages Yearly Compensation")
plt.xlabel("Mean Salary in $")
plt.ylabel("Programming languages")
plt.gca().xaxis.set_major_formatter('${x:1.0f}')
plt.legend()
plt.tight_layout()
plt.show()


#https://www.youtube.com/watch?v=ckjUa76F9Vg&list=PLVs9JdE6ZZh0RBT7lIPxUkcTmAUhMEimn&index=7
```
